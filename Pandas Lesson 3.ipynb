```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pltb
import random
%matplotlib inline
```

```python
def createdataset(number=1):
    output = []
    for i in range(number):
        dateset = pd.date_range('20170101','20201231',freq= 'W-MON')
        status = [1,2,3]
        status_set = [random.choice(status) for i in range(len(dateset))]
        states = ['UP','JK','TG','AP','TN','Tn']
        states_set = [random.choice(states) for i in range(len(dateset))]
        randm_numbers = np.random.randint(100,1000,len(dateset))

        output.extend(zip(dateset,status_set,states_set,randm_numbers))
    return (output)

createdataset()
```

```python
dataset = createdataset()
df = pd.DataFrame(dataset,columns = ['Time','Status','States','Vistors'])
```
