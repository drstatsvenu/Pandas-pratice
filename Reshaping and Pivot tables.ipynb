```python
import pandas as pd
import numpy as np
```

```python
#randomly generated dataframe 
#code copied from pandas file
import pandas.util.testing as tm; tm.N = 3
def unpivot(frame):
    N, K = frame.shape
    data = {'value' : frame.values.ravel('F'),
    'variable' : np.asarray(frame.columns).repeat(N),
    'date' : np.tile(np.asarray(frame.index), K)}
    return pd.DataFrame(data, columns=['date', 'variable', 'value'])
df = unpivot(tm.makeTimeDataFrame())
```

```python
df
```

```python
#to select everything out with variable A
df[df['variable'] == 'A']
```

```python
#to make the above data more readable pivot the date column to index using pivot function
df.pivot(index = 'date' ,columns = 'variable' , values = 'value')
```

```python
#in the above function it doesn't matter even if you dont give values it will automatically select the values
df['values2'] = df['value']*2
```

```python
#pivot_table is used to make spreadsheet type labels
import datetime
df = pd.DataFrame({'A': ['one','two','three','four']*6,
                   'B': ['A','B','C']*8,
                   'C': ['foo','foo','foo','bar','bar','bar']*4,
                   'D': np.random.randn(24),
                   'E': np.random.randn(24),
                   'F': [datetime.datetime(2017,i,1) for i in range(1,13)]+
                        [datetime.datetime(2017,i,1) for i in range(1,13)]})
```

```python
df
```

```python
#well now we have data and from this data we can develop pivot tables pretty much easily
pd.pivot_table(df,index = ['A','B'],columns = 'C',values = ['D','E'])
```

```python
#cross tabulations
foo,bar,dull,shiny,one,two = 'foo','bar','dull','shiny','one','two'
```

```python
a = np.array([foo, foo, bar, bar, foo, foo], dtype=object)
b = np.array([one, one, two, one, two, one], dtype=object)
c = np.array([dull, dull, shiny, dull, dull, shiny], dtype=object)
```

```python
pd.crosstab(a,[b,c],rownames = ['Mr.Row'],colnames = ['Mr.col1','Mr.col2'])
```
